```python
def InputIntOrFloat():
    invalidNumber = True
    number = 0.0
    
    while invalidNumber:
        numberString = input("Type a number ")
        try:
            number = abs(int(numberString))
            invalidNumber = False
        except:
            try:
                number = abs(float(numberString))
                invalidNumber = False
            except:
                print("Please type a valid number without any special characters ")
                invalidNumber = True
    return number
```

```python
def getSelection(message="",maxOption=1):
    selection = 0
    try:
        selection = int(input(message))
        if(selection>maxOption or selection<1):
            print("Option not available, please select an available option")
            return getSelection(message,maxOption)
        else:
            return selection
    except:
        selection = input("Would you like to quit? [q or y]es [N]o\n")
        if(selection.lower().replace(" ","") != 'y' and selection.lower().replace(" ","") != 'q'):
            return getSelection(message,maxOption)
        else:
            quit()
```

```python
class FoodItem:
    
    #Properties
    _notes = ""
    @property
    def Notes(self):
        return self._notes
    @Notes.setter
    def Notes(self,newNotes):
        self._notes = newNotes
    
    _quantity = 1
    @property
    def Quantity(self):
        return self._quantity
    @Quantity.setter
    def Quantity(self,newQuantity):
        self._quantity = newQuantity
        
    _price = 0.0
    @property
    def Price(self):
        return self._price
    
    @Price.setter
    def Price(self,newPrice):
        self._price = newPrice
    
    #Operators
    def __add__(self,other):
        return self.Price + other
    
    def __radd__(self, other):
        if other == 0:
            return self
        else:
            return self.__add__(other)
    
    #Overloads
    def __float__(self):
        return self.Price
        
    #Constructor
    def __init__(self,notes = "", quantity = 1):
        self._notes = notes
        self._quantity = quantity
    
```

```python
class Burger(FoodItem):
    _ingredients = {"LETTUCE":1, "ONION":1,"TOMATO":1,"PICKLES":1,"MUSTARD":1, "CHEESE":1,"BACON":1, "PATTY":1}
    _prices = {"LETTUCE":0.3, "ONION":0.3,"TOMATO":0.2,"PICKLES":0.2,"MUSTARD":0.0, "CHEESE":0.5,"BACON":1.0, "PATTY":1.0}
    
    def ListIngredients(self):
        ingredientsList = list(self._ingredients)
        for i in range(len(ingredientsList)):
            print(f"     {ingredientsList[i]}"+ (30-len(ingredientsList[i]))*" " + f" QTY {self._ingredients[ingredientsList[i]]}")
    def ListExistingIngredients(self):
        ingredientsList = list(self._ingredients)
        for i in range(len(ingredientsList)):
            QTY = self._ingredients[ingredientsList[i]]
            if(QTY>0): print(f"     {ingredientsList[i]}"+ (30-len(ingredientsList[i]))*" " + f" QTY {QTY}")
    
    def ListNumberedIngredients(self):
        ingredientsList = list(self._ingredients)
        for i in range(len(ingredientsList)):
            side = f"[{i+1}] {ingredientsList[i]}"
            print("     "+ side + (30-len(side))*" " + f" QTY {self._ingredients[ingredientsList[i]]}")
    

    
    def _ModifyItems(self):

        print("Which item quantity would you like to modify?")
        self.ListNumberedIngredients()

        lastOption = len(self._ingredients)+1
        print(f"     [{lastOption}] edit ingredient list")
        selection = getSelection(f"Select option from [1] to [{lastOption}] ",lastOption)

        itemName = ""#Maps to the item quantity using _ingredients and the price using _prices

        #When they've chosen to modify the ingredient list itself
        if(selection == lastOption):
            itemName = input("Type a name for the item you'd like to add: ").upper()
            print("Type the quantity of the ingredient you'd like to change to ")
            self._ingredients[itemName] = InputIntOrFloat()
            print("Type the price of the ingredient ")
            self._prices[itemName] = InputIntOrFloat()
        else: #when they're just changing the quantity of an ingredient
            itemName = list(self._ingredients)[selection-1]
            #Now we modify the quantity of the ingredient
            print(f"What would you like the new quantity of {itemName} to be? ")
            self._ingredients[itemName] = InputIntOrFloat()

    def _calculatePrice(self):
        itemList = list(self._ingredients)
        totalCost = 0
    
        for i in itemList: 
            totalCost += self._ingredients[i]*self._prices[i]
        
        return totalCost
    
    def __str__(self):
        print("BURGER:")
        self.ListExistingIngredients()
        return str(f"     ${round(self._price,2)} ")
        
    
    #Constructor
    def __init__(self):
        print("BURGER OPTIONS:")
        self.ListIngredients()
        
        #To see if they want to modify their burger
        while (getSelection("Would you like to modify any items on the burger? [1]Yes [2]No ",2) == 1):
            self._ModifyItems()
            self.ListExistingIngredients()
        
        #Done modifying items, calculating cost of burger
        self._price = self._calculatePrice()
        print(self)        
```

```python

class Drink(FoodItem):
    _size = "MEDIUM"
    @property
    def Size(self):
        return self._size
    
    @Size.setter
    def Size(self,newSize):
        if(newSize=="SMALL"):
            self.Price = 1.0
        elif(newSize == "MEDIUM"):
            self.Price = 1.5
        elif(newSize == "LARGE"):
            self.Price = 2.0
        self._size = newSize
    
    #Operators
    def __str__(self):
        return f"{self.Size} FOUNTAIN DRINK:\n     ${self.Price}"
    
    def __init__(self):
        print("FOUNTAIN DRINK OPTIONS:")
        print("Select a fountain drink size: ")
        selection = getSelection("[1]Small [2]Medium [3]Large ",3)
        if(selection ==1):
            self.Size = "SMALL"
        elif(selection == 2):
            self.Size = "MEDIUM"
        elif(selection == 3):
            self.Size = "LARGE"
        print(self)
        
```

```python
class Side(FoodItem):
    _size = "MEDIUM"
    @property
    def Size(self):
        return self._size
    
    @Size.setter
    def Size(self,newSize):
        if(newSize=="SMALL"):
            self.Price = 1.25
        elif(newSize == "MEDIUM"):
            self.Price = 2.0
        elif(newSize == "LARGE"):
            self.Price = 2.75
        self._size = newSize
    
    _kind = "FRIES"
    @property
    def Kind(self):
        return self._kind
    
    @Kind.setter
    def Kind(self,newKind):
        self._kind = newKind
    
    
    #Operators
    def __str__(self):
        return f"{self.Size} {self.Kind}:\n     ${self.Price}"
    
    def __init__(self):
        print("SIDE OPTIONS:")
        selection = getSelection("[1]Fries [2]Onion Rings [3]Salad ",3)
        if(selection ==1):
            self.Kind = "FRIES"
        elif(selection == 2):
            self.Kind = "ONION RINGS"
        elif(selection == 3):
            self.Kind = "SALAD"       
        
        print(f"Select a {self.Kind} size: ")
        selection = getSelection("[1]Small [2]Medium [3]Large ",3)
        if(selection ==1):
            self.Size = "SMALL"
        elif(selection == 2):
            self.Size = "MEDIUM"
        elif(selection == 3):
            self.Size = "LARGE"
            
        print(self)
```

```python
class Combo(FoodItem):
    _burger = None
    _drink = None
    _side = None
    
    def __str__(self):
        print("COMBO:")
        return f"{self._burger}\n{self._drink}\n{self._side}"
    
    def __init__(self):
        print("COMBO INCLUDES A BURGER, DRINK, AND SIDE:")
        self._burger = Burger()
        self._drink = Drink()
        self._side = Side()
        
        self._price = self._burger.Price + self._drink.Price + self._side.Price
        
```

```python
def take_order():
    #ask user for name for the order 
    #repeat taking order until client is done 
    #display order details
    #display a thank you message
    
    print("Welcome to Burger Shop - press q to quit any time")
    receipt = []
    
    while(getSelection("Continue to order? [1]Yes [2]No ",2)==1):  
        selection = getSelection("What would you like to order?:\n[1]Burger\n[2]Drink\n[3]Side\n[4]Combo\n",maxOption = 4)
        if(selection==1):
            receipt.append(Burger())
        elif(selection==2):
            receipt.append(Drink())
        elif(selection==3):
            receipt.append(Side())
        elif(selection==4):
            receipt.append(Combo())
        else:
            print("This should never ever execute")
            raise Exception('FATAL LOGIC ERROR')
        
    #Printing the receipt items
    print("\n-------------------------------------------\nRECEIPT\n")
    totalPrice = 0
    for i in receipt:
        totalPrice += i.Price
        print(i,"\n")
    
    print(f"-------------------------------------------\n\nYOUR ORDER TOTAL IS ${totalPrice}")
    #print(f"-------------------------------------------\n\nYOUR ORDER TOTAL IS ${sum(receipt)}") #Fails to convert to float when summing exactly one Drink
    
take_order()
```

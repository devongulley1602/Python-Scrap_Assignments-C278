```python
m = [[' _ ',' _ ',' _ ',' _ ',' _ '],[' _ ',' _ ',' _ ',' _ ',' _ '],[' _ ',' _ ',' _ ',' _ ',' _ ']]

# for i in range(len(m)):
#     for j in range(len(m[i])):
#         m[i][j] = "("+str(i) + "," +  str(j)+")" 

def printBoard():
    for i in range(len(m)):
        print(*m[i])
    print()

def checkWinner():
    #Diagonal win conditions
    rd = (m[0][0] == m[1][1]) and (m[1][1] == m[2][2]) and (m[0][0] != ' _ ')
    ld = (m[0][2] == m[1][1]) and (m[1][1] == m[2][0]) and (m[0][2] != ' _ ')
    
    #Row and Column win conditions
    for i in range(len(m)):
        rw = (m[i][0] == m[i][1]) and (m[i][1] == m[i][2]) and (m[i][0] != ' _ ')
        cw = (m[0][i] == m[1][i]) and (m[1][i] == m[2][i]) and (m[0][i] != ' _ ')
        if(rw or cw):
            break
    return rw or cw or rd or ld
        
turns = 0
winCondition = False
drawCondition = False

while not (winCondition or drawCondition):
    selected = False
    token = ''
    while not selected:
        printBoard()
        #For whomever's turn it is
        if turns%2 == 0:
            token = ' x '
        else:
            token = ' o '
        
        print(f"It is {token}'s turn...")
        
        
        row = int(input("Select a row [0 to 2]: "))
        col = int(input("Select a column [0 to 2]:  "))
        selected = ((m[row][col] == ' _ ') and (row <= 2) and (col <=2) and(row>=0) and (col>=0))
        
        if(selected):
            turns += 1
            m[row][col] = token
        else:
            print("Invalid input, please try again")
    winCondition = checkWinner()
    drawCondition = (not winCondition) and (turns >= 9)

printBoard()

if not drawCondition:
    print(f"{token} won")
else:
    print("draw")
        
```

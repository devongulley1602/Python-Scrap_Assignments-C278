```python
proceed = False
price = 0.0

while not proceed:
    size = input("What size would you like? [s]mall [M]edium [l]arge\n")
    
    #Formatting input, calculating cost
    size = size.upper().replace(" ","")
    
    if ((size == "" )| (size == "M") | (size == "MED") | (size == "MEDIUM")): 
        size = "MEDIUM"
        price = 3   
    elif ((size == "S") | (size == "sml") | (size == "SMALL")) : 
        size = "SMALL"
        price = 2   
    elif ((size == "LRG") | (size == "L") | (size == "LARGE")): 
        size = "LARGE"
        price = 4        
    
    price = round(price,2)
    proceed = ((size == "SMALL") | (size == "MEDIUM") | (size == "LARGE"))
    
    
print(f"You have selected size {size} for a cost of ${price}\n")
proceed = False;



while not proceed:
    brewType = input("Which brew would you like? [e]spresso [B]rewed [c]old_brew\n")
    
    #Formatting input and calculating additional costs
    brewType = brewType.upper().replace(" ","")
    
    if ((brewType == "")| (brewType == "B") | (brewType == "BREWED")):
        brewType = "BREWED";
    elif ((brewType == "E") | (brewType == "ESP") | (brewType == "ESPRESSO") | (brewType == "EXP") | (brewType == "EXPRESSO")):
        brewType = "ESPRESSO";
    elif ((brewType == "C")|(brewType == "CB")| (brewType == "COLDBREW") | (brewType == "COLD_BREW")):
        brewType = "COLD_BREW";
    
    proceed = ((brewType == "COLD_BREW") | (brewType == "ESPRESSO") | (brewType == "BREWED"))
    
print(f"You have selected {brewType}\n")
proceed = False;

flavour = input("Which flavour would you like? [v]anilla [h]azelnut [c]aramel [N]one\n")
flavour = flavour.upper().replace(" ","")

#Formatting for receipt
if (flavour == "V") | (flavour == "VANILLA") | (flavour == "VAN"):
    flavour = "VANILLA";
elif (flavour == "H") | (flavour == "HZLNT") | (flavour == "HAZELNUT") | (flavour == "HAZLENUT"):
    flavour = "HAZELNUT";
elif (flavour == "C") | (flavour == "CAR") | (flavour == "CARAMEL") | (flavour == "CARMEL"):
    flavour = "CARAMEL";
else:
    flavour = "NONE";

if flavour != "NONE":
    price += 0.5;

print(f"Receipt:\n\n    Brew: {brewType}\n    Size: {size}\n    Flavour: {flavour} \n    Price (+15% tip): ${round(price*1.15,2)}")


    
        

    
```
